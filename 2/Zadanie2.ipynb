```python
import pandas as pd
```

```python
data = pd.read_excel('QTL_Sample_data.xls')
data
```

```python
dimensions = data.shape
print("Liczba wierszy i kolumn: ", dimensions)
```

```python
data_mean=data["age"].mean()
print("Średni wiek:", data_mean)
```

```python
data_max=data["brainwt"].max()
print("Najwyższa wartość brainwt:", data_max)
```

```python
data_min=data["brainwt"].min()
print("Najmniejsza wartość brainwt:", data_min)
```

```python
plec = data[data['ID']==1709].copy()
print("Płeć myszy o ID 1709:", plec.iloc[0]['sex'])
```

```python
ID_max = data[data['bodywt'] == data['bodywt'].max()].copy()
print("ID myszy o najwyższej wartości bodywt:", ID_max.iloc[0]['ID'])
```

```python
sex_count = data['sex'].value_counts().copy()
print(sex_count)
```

```python
data['sex']=data['sex'].str.replace(' ','').copy()
```

```python
sex_count = data['sex'].value_counts().copy()
print("Ilość myszy płci męskiej i żeńskiej:")
print(sex_count)
```

```python
type_count = data['Strain'].value_counts().copy()
print("Rodzajów myszy jest:", type_count.count())
```

```python
brainwt_m = data[data['sex']=="M"].copy()
brainwt_m_mean = brainwt_m['brainwt'].mean()
brainwt_f = data[data['sex']=="F"].copy()
brainwt_f_mean = brainwt_f['brainwt'].mean()
bodywt_m = data[data['sex']=="M"].copy()
bodywt_m_mean = bodywt_m['bodywt'].mean()
bodywt_f = data[data['sex']=="F"].copy()
bodywt_f_mean = bodywt_f['bodywt'].mean()
print("Średnia wartość brainwt dla myszy męskich:", brainwt_m_mean)
print("Średnia wartość brainwt dla myszy żeńskich:", brainwt_f_mean)
print("Średnia wartość bodywt dla myszy męskich:", bodywt_m_mean)
print("Średnia wartość bodywt dla myszy żeńskich:", bodywt_f_mean)
```
