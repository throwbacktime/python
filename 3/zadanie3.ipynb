```python
import matplotlib.pyplot as plt
import matplotlib.patches as mp
import pandas as pd
```

```python
data = pd.read_csv('zwiazki.out', sep='\s+' , usecols=[1,2,3,4,5], skiprows=3, names=['CO', 'CO2', 'O2', 'NO', 'THC'])
data
```

```python
fig, ax1 = plt.subplots(figsize=(8,8))

ax1.set_xlabel('Iteracje [-]', fontsize=10)
ax1.set_ylabel('$CO_2$, $O_2$ [%], $CO$ [ppm]', fontsize=10)
ax1.plot(data['CO'], 'blue', label='$CO$',linewidth=3)
ax1.plot(data['CO2'], 'green', label='$CO_2$',linewidth=3)
ax1.plot(data['O2'], 'red', label='$O_2$',linewidth=3)
ax1.set_ylim(0,40)
ax1.set_xlim(0,1400)
plt.grid(True)

ax2 = ax1.twinx()
ax2.set_ylabel('$THC$, $NO$ [ppm]', fontsize=10)
ax2.plot(data['THC'], 'orange', label='$THC$',linewidth=3)
ax2.plot(data['NO'], 'black', label='$NO$',linewidth=3)
ax2.set_yscale('log')
ax2.set_ylim(10e0, 10e4)
plt.title ('SkÅ‚adniki spalin')
fig.legend(bbox_to_anchor=(0.2,1), bbox_transform=ax1.transAxes)
plt.show()
```
