```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('pix_diff1.txt')
data2 = pd.read_csv('pix_diff2.txt')
data3 = pd.read_csv('pix_diff3.txt')
data4 = pd.read_csv('pix_diff4.txt')
```

```python
czas1 = len(data1)/480
czas2 = len(data2)/480
czas3 = len(data3)/480
czas4 = len(data4)/480
```

```python
droga1 = (data1.iloc[:,0].max()-data1.iloc[:,0].min())/7
droga2 = (data2.iloc[:,0].max()-data2.iloc[:,0].min())/7
droga3 = (data3.iloc[:,0].max()-data3.iloc[:,0].min())/7
droga4 = (data4.iloc[:,0].max()-data4.iloc[:,0].min())/7

predkosc1 = droga1/czas1
predkosc2 = droga2/czas2
predkosc3 = droga3/czas3
predkosc4 = droga4/czas4
```

```python
print("Droga w pierwszym przypadku:", round(droga1, 2),"[mm], predkosc:", round(predkosc1, 2),"[mm/s] oraz czas:", round(czas1, 3),"[s]")
print(" ")
print("Droga w drugim przypadku:", round(droga2, 2),"[mm], predkosc:", round(predkosc2, 2),"[mm/s] oraz czas:", round(czas2, 3),"[s]")
print(" ")
print("Droga w trzecim przypadku:", round(droga3, 2),"[mm], predkosc:", round(predkosc3, 2),"[mm/s] oraz czas:", round(czas3, 3),"[s]")
print(" ")
print("Droga w czwartym przypadku:", round(droga4, 2),"[mm], predkosc:", round(predkosc4, 2),"[mm/s] oraz czas:", round(czas4, 3),"[s]")
```
