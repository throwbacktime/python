```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=int(input("Podaj parametr a: "))
```

```python
b=int(input("Podaj parametr b: "))
```

```python
c=int(input("Podaj parametr c: "))
```

```python
delta = (b*b) - (4*a*c)
print("Delta = ", delta)
if delta==0:
    x0 = (-b)/(2*a)
    print("Funkcja ma 1 miejsce zerowe: ",x0)
elif delta >0:
    x1 = (-b-math.sqrt(delta))/(2*a)
    x2 = (-b+math.sqrt(delta))/(2*a)
    print("Funkcja ma 2 miejsca zerowe: " ,round(x1,2) ,"i" ,round(x2,2))
else:
    print("Delta mniejsza od zera - nie ma rozwiązań")

x_e = -b/(2*a)
y_e = a*x_e*x_e + b*x_e + c
print("Ekstremum wynosi: " ,y_e ,"dla x równego: " ,x_e)
```

```python
m=[]
n=[]
for x in range (-5,5,1):
        y=a*(x**2)+b*x+c
        m.append(x)
        n.append(y)
        
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(m,n, label='wykres f. kwadratowej')
axes.plot
plt.scatter([x_e], [y_e], label='ekstremum')
if delta >0:
    plt.scatter([x1, x2], [0, 0], label='m. zerowe')
plt.title("Wykres f. kwadratowej", fontsize=15) 
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
