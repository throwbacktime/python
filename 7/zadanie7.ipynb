```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_excel('impulse_100_90_85_80_70_mod.xlsx')
data2 = pd.read_excel('impulse_95_75_65_60_55_mod.xlsx')
data3 = pd.read_excel('pelton_all_mod.xlsx')
data4 = pd.read_excel('reaction_100_90_85_80_70_mod.xlsx')
data5 = pd.read_excel('reaction_95_75_65_60_55_mod.xlsx')
data1.drop(data1.columns[0], axis=1, inplace=True)
data2.drop(data2.columns[0], axis=1, inplace=True)
data3.drop(data3.columns[0], axis=1, inplace=True)
data4.drop(data4.columns[0], axis=1, inplace=True)
data5.drop(data5.columns[0], axis=1, inplace=True)
```

```python
data1[data1<0]=0
data2[data2<0]=0
data3[data3<0]=0
data4[data4<0]=0
data5[data5<0]=0
file=data1
file
```

```python
e=[]
f=[]
g=[]
h=[]
hh=[]
mm=[]
mn=[]
tytul=[]
for i in range(len(file)):
    if (file.iloc[i][5]==100):
        e.append(data1.iloc[i][5])

    elif (file.iloc[i][5]==90):
        f.append(data1.iloc[i][5])

    elif (file.iloc[i][5]==85):
        g.append(data1.iloc[i][5])

    elif (file.iloc[i][5]==80):
        h.append(data1.iloc[i][5])

    elif (file.iloc[i][5]==70):
        hh.append(data1.iloc[i][5])

tytul.append('100%')
tytul.append('90%')
tytul.append('85%')
tytul.append('80%')
tytul.append('70%')

mm.append(0)
mn.append(len(e))
mm.append(len(e))
mn.append(len(e)+len(f))
mm.append(len(e)+len(f))
mn.append(len(e)+len(f)+len(g))
mm.append(len(e)+len(f)+len(g))
mn.append(len(e)+len(f)+len(g)+len(h))        
mm.append(len(e)+len(f)+len(g)+len(h))
mn.append(len(e)+len(f)+len(g)+len(h)+len(hh))
```

```python
for k in range(5):
    op=mm[k]
    oo=mn[k]
        
    y1=[]
    x1=[]
    x2=[]
    for j in range (op, oo):
        w=2*3.14/60*file.iloc[j,3]
        m=file.iloc[j,4]*0.045
        pt=m*w
        pp=file.iloc[j,1]/60*file.iloc[j,2]
        n=pt/pp*100
        if (n>0):
            y1.append(file.iloc[j,3])
            x1.append(n)
            x2.append(pt)
    plt.figure()
    plt.plot(y1,x1, label='Sprawno')
    plt.plot(y1,x2, label='Moc turbiny')
    plt.title(tytul[k], fontsize=15) 
    plt.xlabel('Prdko obrotowa RPM')
    plt.ylabel('n/pt')
    plt.legend()
plt.show()
```
