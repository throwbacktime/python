```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('data', delimiter = ',', usecols = [0,1,12,10,11])
wykres=['', '','Liczba Macha \nmin =%5.2f \nmax =%5.2f \n'%(data[:,2].min(), data[:,2].max()), 'Cinienie [Pa] \nmin =%5.2f \nmax =%5.2f \n'%(data[:,3].min(), data[:,3].max()), 'Temperatura [K] \nmin =%5.2f \nmax =%5.2f \n'%(data[:,4].min(), data[:,4].max())]
nazwa=['','','mach','cinienie','temperatura']
for i in range(2,5):
    plt.figure(figsize=(14.2,3.5), dpi=600)
    plt.tricontourf(data[:,0], data[:,1], data[:,i], 100)
    plt.annotate(wykres[i], xy=(0.03, 0.7), xycoords='axes fraction', bbox=dict(boxstyle="square,pad=0.3", fc="white", ec="b",))
    plt.colorbar()
    plt.clim(data[:,i].min(), data[:,i].max())
    plt.ylim(0,4)
    plt.xlabel("wsprzdne x")
    plt.ylabel("wsprzdne y")
    #plt.savefig((nazwa[i]))
```

```python
x=[]
y=[]
z=[]
for j in range(len(data)):
    if (data[j,1]==0):
        x.append(data[j,0])
        y.append(data[j,2])
        z.append(data[j,3])
        
fig, ax1 = plt.subplots(dpi=600)
ax1.set_xlabel('wsprzdne x')
ax1.set_ylabel('liczba Macha')
ax1.plot(x,y, color='red', label='liczba Macha')
ax2 = ax1.twinx() #inicjalizacja drugiej osi

ax2.set_ylabel('cinienie')
ax2.plot(x, z, color='blue', label='cinienie')
fig.legend(loc='center')
plt.show()
```
