```python
import matplotlib.pyplot as plt
import matplotlib.patches as mp
import pandas as pd
from matplotlib import cm, colors
import numpy as np
```

```python
data = pd.read_csv('cross-section.dat')
head_list = data.columns.tolist()
print(head_list)
```

```python
x = data[head_list[1]].values.tolist()
sym_x = [i*(-1) for i in x]
x = x+sym_x

y = data[head_list[2]].values.tolist()
y = y+y

o2_dry = data[head_list[6]].copy()
o2_dry[o2_dry>15]=15
o2_dry = o2_dry.tolist()
o2_dry = o2_dry + o2_dry

no_dry = data[head_list[5]].copy()
no_dry[no_dry>30000]=30000
no_dry = no_dry.tolist()
no_dry = no_dry + no_dry

co_dry = data[head_list[4]].copy()
co_dry[co_dry>100000]=100000
co_dry = co_dry.tolist()
co_dry = co_dry + co_dry

co2_dry = data[head_list[7]].copy()
co2_dry[co2_dry>20]=20
co2_dry = co2_dry.tolist()
co2_dry = co2_dry + co2_dry

v = data[head_list[8]].copy()
v[v>100]=100
v[v<5]=5
v = v.tolist()
v = v + v
```

```python
plt.figure(figsize=(40,30))

plt.subplot(2,3,1)
plt.scatter(x, y, s=30, c=o2_dry, marker='o', cmap=cm.jet)
plt.colorbar()
plt.title ('$o2-dry$', fontsize=15)
plt.xlabel('X', fontsize=8)
plt.ylabel('Y', fontsize=8)

plt.subplot(2,3,2)
plt.scatter(x, y, s=30, c=no_dry, marker='o', cmap=cm.jet)
plt.colorbar()
plt.title ('$no-ppm-dry$', fontsize=15)
plt.xlabel('X', fontsize=8)
plt.ylabel('Y', fontsize=8)

plt.subplot(2,3,3)
plt.scatter(x, y, s=30, c=co_dry, marker='o', cmap=cm.jet, norm=colors.LogNorm())
plt.colorbar() 
plt.title ('$co-ppm-dry$', fontsize=15)
plt.xlabel('X', fontsize=8)
plt.ylabel('Y', fontsize=8)

plt.subplot(2,3,4)
plt.scatter(x, y, s=30, c=co2_dry, marker='o', cmap=cm.jet)
plt.colorbar()
plt.title ('$co2-dry$', fontsize=15)
plt.xlabel('X', fontsize=8)
plt.ylabel('Y', fontsize=8)

plt.subplot(2,3,5)
plt.scatter(x, y, s=30, c=v, marker='o', cmap=cm.jet)
plt.colorbar()
plt.title ('$velocity-magnitude$', fontsize=15)
plt.xlabel('X', fontsize=8)
plt.ylabel('Y', fontsize=8)

plt.show()
```
