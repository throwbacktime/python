```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path_list = ['./p1', './p4','./p5']

o2_p1=[]
o2_p4=[]
o2_p5=[]

for j in range(0,3):
    path=path_list[j]
    file_list= os.listdir(path)
    o2_srednia = np.zeros(len(file_list))
    i=0
    for number in file_list:
        data = pd.read_csv(path+r"/"+number ,skiprows = 23, sep="\s+", usecols=[16], names=["o2"], decimal=",")    
        if (j==0):
            o2_p1.append(data["o2"].tail(15).mean())
        elif (j==1):
            o2_p4.append(data["o2"].tail(15).mean())
        elif (j==2):
            o2_p5.append(data["o2"].tail(15).mean())
        i=i+1

x_p1 = np.arange(-175,175,24)
x_p4_5 = np.arange(-210,210,29)

plt.figure()
plt.plot(x_p1, o2_p1, label="p1")
plt.plot(x_p4_5, o2_p4, label="p4")
plt.plot(x_p4_5, o2_p5, label="p5")
plt.xlabel('x [mm]')
plt.ylabel('o2 [%]')
plt.grid(True)
plt.legend()
plt.show()
```
